```python
def calcular_media(notas):
    if len(notas) == 0:
        return 0
    soma = sum(notas)
    media = soma / len(notas)
    return media

def exibir_alunos(alunos):
    print("\nAlunos e suas notas:")
    for aluno, notas in alunos.items():
        print(f"{aluno}: {notas}")
    print()

def main():
    alunos = {}

    while True:
        nome = input("Digite o nome do aluno (ou 'parar' para encerrar): ")
        if nome.lower() == 'parar':
            break

        notas = []
        while True:
            nota_str = input(f"Digite a nota para {nome} (ou 'parar' para encerrar): ")
            if nota_str.lower() == 'parar':
                break
            nota = float(nota_str)
            notas.append(nota)

        alunos[nome] = notas

    exibir_alunos(alunos)
    media_geral = calcular_media([nota for notas in alunos.values() for nota in notas])
    print(f"MÃ©dia geral das notas: {media_geral:.2f}")

if __name__ == "__main__":
    main()
```
