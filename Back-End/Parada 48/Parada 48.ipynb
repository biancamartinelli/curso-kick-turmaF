```python
meses = ['janeiro', ' fevereiro', 'mar√ßo', 'abril', 'maio']
vendas = [15000, 22000, 18000, 20000, 40000]

plt.plot(meses, vendas, marker='o', color='b', linestyle='--', label='vendas')
plt.title('Teste')
plt.xlabel('meses')
plt.ylabel('vendas')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

idades = [25, 35, 40, 47, 55]
salario = [1300, 1500, 2400, 4500, 6000]

scatter_fig = px.scatter(x=idades, y=salario, title ='rela√ß√£o entre idade e sal√°rio')
scatter_fig.update_xaxes(title_text='x')
scatter_fig.update_yaxes(title_text='y')
scatter_fig.show()
```

```python
import plotly.graph_objs as go
import numpy as np

anos = np.arange(2010, 2023)
receita = np.random.randint(50000, 80000, size=len(anos))

trace = go.Scatter(x=anos, y=receita, mode='lines+markers')

layout = go.Layout(
    title='Receita ao longo dos anos',
    xaxis=dict(title='ano'),
    yaxis=dict(title='receita'),
    sliders=[{
        'steps': [{'args': [['x', 'y'], [anos[:i+1]], receita[:i+1]],
                   'label': str(ano),
                   'method': 'restyle'} for i, ano in enumerate(anos)],
        'active': 0
    }]
)

fig = go.Figure(data=[trace], layout=layout)
fig.show()
```
